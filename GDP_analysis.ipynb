```python
#Analysis Part I-A
#importing .csv file GSDP_Data

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

#Converting the .csv file into a dataframe 
df = pd.read_csv("G:/upGrad+iiitb/GDP_Analysis/Data_IA/GSDP_Data.csv", encoding = "ISO-8859-1")

#Removing the data for the year 2016-17

df1 = df[df.Duration != '2016-17']
#df1.head()

#Taking the mean values of the state
df2 = df1.tail(3).mean()

#df2

#Plotting the avg. growth of states

avg_gth = df2.plot(kind = 'bar', stacked = True, figsize = (15,6), x = 'States', y = 'Mean')
avg_gth.set_xlabel("State")
avg_gth.set_ylabel("Average Growth")
plt.show()

# Plotting total GDP of states for the year 2015-16

df3 = df1.loc[(df1.Duration == '2015-16'), :]
GDP_1516 = df3.iloc[0,2:-1].plot(kind = 'bar', figsize = (15,6), x= 'States', y = 'GDP')
GDP_1516.set_xlabel("State")
GDP_1516.set_ylabel("Total GDP")
plt.show()
```
